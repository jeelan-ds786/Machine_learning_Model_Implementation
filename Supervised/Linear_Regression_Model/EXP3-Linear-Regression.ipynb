# Linear Regression - Univariate

### Import module and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/sanja/Downloads/Salary_Data.csv")
df.head()
```

### Differentiating Dependend and Independed variables

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```

### Fitting into Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(xtrain, ytrain)
```

```python
reg.coef_, reg.intercept_
```

```python
ypred = reg.predict(xtest)
xpred = reg.predict(xtrain)
ypred, xpred
```

### Visualizing

```python
ypred = reg.predict(df[['YearsExperience']])
plt.scatter(df.YearsExperience, df.Salary)
plt.plot(df.YearsExperience,ypred , color='red')
```

```python
ypred
```

```python
7.5/4
```

```python
ypred1 = (df.Salary-min(df.Salary))/(max(df.Salary)-min(df.Salary))
x, ypred1
```

```python
ypred2 = (ypred-min(ypred))/(max(ypred)-min(ypred))
x, ypred2
```

```python
plt.scatter(x, ypred1)
plt.plot(df.YearsExperience,ypred2 , color='red')
plt.grid(True)
```

### Determing Accuracy

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, ypred)**0.5
```

```python
mean_squared_error(ypred2, ypred1)**0.5
```
