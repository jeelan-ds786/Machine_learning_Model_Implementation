```python
df['zscoreProfit'] = (df.Profit - df.Profit.mean())/df.Profit.std()
df
```

```python
df = df[(df.zscoreProfit<2) & (df.zscoreProfit>-2)]
df
```

```python
df.Performance.unique()
```

```python
d = {'HIGH': 2, 'MEDIUM' : 1, 'LOW' : 0}
df.Performance = df.Performance.map(d)
df
```

```python
dfPerformance = df.Performance
dfPerformance
```

```python
df.columns
```

```python
df =df.drop(['State', 'Total Earnings', 'Performance', 'zscoreProfit'], axis = 1)
df
```

```python
sns.pairplot(df)
```

```python
x = df.iloc[:,2:]
x
```

```python
y = dfPerformance
y = pd.DataFrame(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(xtrain, ytrain)
```

```python
knn.score(xtest, ytest)
```

```python
score = []
for i in range (1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain, ytrain)
    sc = knn.score(xtest, ytest)
    score.append(sc)
score
```

```python
plt.plot(range(1,20), score)
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import confusion_matrix
ypred = knn.predict(xtest)
cm = confusion_matrix(ytest, ypred)
cm
```

```python
ypred = knn.predict(x)
cm = confusion_matrix(y, ypred)
cm
```

```python
plt.scatter(x[ypred ==0]["Marketing Spend"], x[ypred == 0]['Profit'],  
            c='purple', label = 'LOW')
plt.scatter(x[ypred ==1]["Marketing Spend"], x[ypred == 1]['Profit'],  
            c='orange', label = 'MEDIUM')
plt.scatter(x[ypred ==2]["Marketing Spend"], x[ypred == 2]['Profit'], 
            c='green', label = 'HIGH')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.legend()
```
