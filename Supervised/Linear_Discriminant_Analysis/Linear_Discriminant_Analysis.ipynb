## LDA - Multiclass

```python
db = pd.read_csv("C:/Users/sanja/Downloads/DryBeanDataset/DryBeanDataset/Dry_Bean_Dataset.csv")
db
```

```python
db.Class.nunique(), db.Class.unique()
```

```python
d = {'SEKER': 0, 'BARBUNYA': 1, 'BOMBAY': 2, 'CALI' :3, 'HOROZ': 4, 'SIRA': 5, 'DERMASON': 6}
db.Class = db.Class.map(d)
db
```

# Splitting the Feature and Target

```python
x = db.iloc[:, :-1]
y = db.iloc[:, -1]
x.shape, y.shape
```

## Splitting the Feature and Target into Train and Test

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

## Fitting into the Model

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=6)
xtrainlda = lda.fit_transform(xtrain, ytrain)
xtestlda = lda.transform(xtest)
pd.DataFrame(xtrainlda)
```

```python
Classes = ['SEKER', 'BARBUNYA', 'BOMBAY', 'CALI', 'HOROZ', 'SIRA', 'DERMASON']
xlda = lda.transform(x)
ypred = lda.predict(x)
for i in range (db.Class.nunique()):
    plt.scatter(xlda[:,0][ypred==i], xlda[:,1][ypred==i], label = Classes[i])
plt.legend()
```

## Predict the test result and validate the score

```python
ypred = lda.predict(xtest)
accuracy_score(ytest, ypred)
```
